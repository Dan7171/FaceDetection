```python
import pandas as pd
import pickle as pkl
```

```python
with open('/home/administrator/experiments/high_importance_features_frontal_faces/vgg16/results/comparisons.pkl', 'rb') as f:
    list_type_to_comparisons = pkl.load(f)
```

```python
list_type_to_comparisons
```

```python
q = None
for list_type in list_type_to_comparisons:
    p = pd.DataFrame(list_type_to_comparisons[list_type]).transpose()
    p['list_type'] = list_type
    print(p)
    if q is None:
        q = p
    else:
        q = pd.concat([q,p])
```

```python
q = q.rename(columns={7:'conv1', 12:'conv2', 19:'conv3', 26:'conv4', 33:'conv5', 37: 'fc6', 40:'fc7', 42:'fc8'})
```

```python
q = q.drop(columns=['output'])
```

```python
q = q[['list_type', 'input', 'conv1', 'conv2', 'conv3', 'conv4', 'conv5', 'fc6', 'fc7', 'fc8']]
```
